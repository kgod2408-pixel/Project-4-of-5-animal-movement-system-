```python
class Animal:
    def __init__(self, name):
        self.name = name

    def move(self):
        print(self.name, "is moving")

class Dog(Animal):
    def move(self):
        print(self.name, "is running on land")

class Bird(Animal):
    def move(self):
        print(self.name, "is flying in the sky")

class Fish(Animal):
    def move(self):
        print(self.name, "is swimming in water")

class Snake(Animal):
    def move(self):
        print(self.name, "is slithering on the ground")
```

```python
dog = Dog("Dog")
bird = Bird("Bird")
fish = Fish("Fish")
snake = Snake("Snake")

dog.move()
bird.move()
fish.move()
snake.move()
```
